# Task 1 - Plot a line graph to show the difference between heart rate of smokers and non smokers

```python
#Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
#read the csv
df = pd.read_csv("framingham.csv")
df
```

```python
#Filter and make a new dataframe for non smokers
NonSmoke = df.loc[df["currentSmoker"] == 0]
NonSmoke
```

```python
#Group by age column and find average heart rate at different age
GroupByAge = NonSmoke.groupby("age")["heartRate"].mean().reset_index()
GroupByAge
```

```python
#Filter and make a new dataframe for smokers
smoke = df.loc[df["currentSmoker"] == 1]
smoke
```

```python
#Group by age column and find average heart rate at different age
GroupByAgeS = smoke.groupby("age")["heartRate"].mean().reset_index()
GroupByAgeS
```

```python
#Plot a line graph to show the heart rate of smokers vs non smokers
fig = plt.figure(figsize=(12,8))

NsLabel = GroupByAge["age"]
NsValue = GroupByAge["heartRate"]

SLabel= GroupByAgeS["age"]
SValue= GroupByAgeS["heartRate"]

plt.plot(NsLabel, NsValue, label="Non Smoker Heart Rate", linewidth=3.0)

plt.plot(SLabel, SValue, label="Smoker Heart Rate", linewidth=3.0)

plt.xticks(rotation="vertical")
plt.xlabel("Age")

plt.title("Non-Smoker and Smoker Heart Rate Comparison", fontsize=20)

plt.legend()

plt.show()
```

# Task 2 - Which age group have high chances of having coronary heart disease in 10 years

```python
#Read the csv

#Filter and make a new dataframe for those who has chances of having coronary heart disease in 10 years
Coronary = df.loc[df["TenYearCHD"] == 1]
Coronary
```

```python
#Group by age column and count the rows of TenYearCHD column
groupByAgeC = Coronary.groupby("age")["TenYearCHD"].count().reset_index()
groupByAgeC
```

```python
#Plot a line graph to show total number of people having a chance of coronary heart disease in 10 years
figsize = plt.subplots(figsize=(12,8))

label = groupByAgeC["age"]
value = groupByAgeC["TenYearCHD"]

plt.plot(label, value, label="Risk of Coronary", linewidth=3.0)

plt.xlabel("Age")
plt.ylabel("TenYearCHD")

plt.title("Risk of Coronary Heart Disease", fontsize=20)

plt.legend()

plt.show()
```
